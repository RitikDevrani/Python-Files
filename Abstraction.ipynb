# Abstraction in Python
- Abstraction is used to hide the implementation details of method from the user.
- The user is familiar with what the method does, but does not know how it does it.
- Abstraction is implemented by creating an Abstract Base Class(ABC) that contains abstract method.
- Abstract methods do not have any implementation.
- All implementaiton are performed within the subclass.
- All subclass of an abstract base class are required to implement all abstract methods.
- You can not create an object of an abstract base class.

```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    def display(self):
        print("This is not abstract method")
```

```python
s = Shape() # can not create object of abstract class
```

```python
class Rectangle(Shape):
    def area(self, length, width):
        return length*width

    def perimeter(self, length, width):
        return length*width
```

```python
r = Rectangle()
```

```python
print(r.area(4,7))
print(r.perimeter(8,2))
r.display()
```

```python
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14*self.radius**2

    def perimeter(self):
        return 2*3.14*self.radius
```

```python
cir = Circle(3)
```

```python
print(cir.area())
cir.perimeter()
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def start(self):
        print("Vehicle started")

    @abstractmethod
    def speed(self):
        pass

class Car(Vehicle):
    def speed(self):
        print("Car speed is 120 km/h")

c = Car()
c.start()
c.speed()
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def draw(self):
        pass

class Square(Shape):
    def area(self):
        print("Area of Square")

    def draw(self):
        print("Drawing Square")

s = Square()
s.area()
s.draw()
```

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def pay(self):
        pass

class UPI(Payment):
    def pay(self):
        print("Paid via UPI")

class Card(Payment):
    def pay(self):
        print("Paid via Card")

payments = [UPI(), Card()]
for p in payments:
    p.pay()
```

# Abstraction with Constructor

```python
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def salary(self):
        pass

class Manager(Employee):
    def salary(self):
        print("Manager salary")

m = Manager("Ritik")
print(m.name)
m.salary()
```

# Python Built-in Functions
- Built-in functions are predefined functions in Python that can be used without importing any module. They perform common tasks like mathematical operations, data type conversions, and input/output operation.

# Common Built-in Functions

```python
# print() - Displays output on the screen
name = "My name is Ritik"
print(name)
```

```python
# input() - Takes input from the user
name1 = input("Enter your name")
age = int(input("Enter your age"))
```

```python
# len() - Returns the length 
len(name)
```

```python
# type() - Returns the data type of the variable
my_list = [4, "riot" , 3.5 , True]
type(my_list)
```

```python
# max() - Returns the largest value
nums = (4, 54, 25, 85, 5.5, 52, 33)
max(nums)
```

```python
# min() - Returns the smallest value
min(nums)
```

```python
# sum() - Returs the total of all elements
sum(nums)
```

```python
# pow() - Ruturns the power of a number
pow(2,3) #(base, power)
```

```python
pow(5.34, 7)
```

```python
pow(5.34, 7.5)
```

```python
# abs()- Returns the absolute(positive) value
a = -577
print(abs(a))
print(abs(-345))
```

```python
# round() - Rounds a number to neares integer
round(88.34525,3)
```

```python
round(88.34525,2)
```

```python
round(88.34525,4)
```

```python
round(88.345257,5)
```

```python
# sorted() - Sorts a list or tuple
my_tuple = (3, 5, 2, 98, 2, 3.5)
print(sorted(my_tuple))
my_listt = [3, 5, 2, 98, 2, 3.5, 45, 99]
print(sorted(my_listt))
```

```python
# rang() - Generates a sequence of number
b = list(range(0,20)) #(startin,ending)
print(b)
c = tuple(range(0,51,5)) ##(startin,ending, stride)
print(c)
```

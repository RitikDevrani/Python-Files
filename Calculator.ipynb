```python
num_1 = input("Enter First Number")

num_2 = input("Enter Second Number")

print("Enter Operation Type\n+\n-\n*\n/")
op_type = input()

if num_1.isdigit() == True and num_2.isdigit() == True:
    num1 = float(num_1)
    num2 = float(num_2)
    if op_type == "+":
        print(num1+num2)
    elif op_type == "-":
        print(num1-num2)
    elif op_type == "*":
        print(num1*num2)
    elif op_type == "/":
        print(num1/num2)
    else:
        print("Plese write valid input")
else:
    print("Please write only numeric values")
```

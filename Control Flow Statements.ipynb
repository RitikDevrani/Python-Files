# Control Flow Statements in Python
- Control Flow Statements Python ko batate hain ki program ka flow kaise chalega — yani kis condition par kaun-sa code chalega.

  - Python me 3 types of control flow hote hain:
   
      - 1️ Conditional Statements (Decision Making)
      - 2️ Looping Statements (Repetition)
      - 3️ Jump Statements (Flow ko break/skip/return karna)

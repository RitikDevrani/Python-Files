```python
# intersection, differentiation, neither
arr6 = np.array([1,2,3,4,4,5,6,7])
arr7 = np.array([5,6,7,8,4,3])
np.intersect1d(arr6, arr7)
```

```python
arr6 = np.array([1,2,3,4,4,5,6,7])
arr7 = np.array([5,6,7,8,4,3])
np.union1d(arr6, arr7)
```

```python
arr6 = np.array([1,2,3,4,4,5,6,7])
arr7 = np.array([5,6,7,8,4,3])
np.setdiff1d(arr6, arr7)
```

```python
arr6 = np.array([1,2,3,4,4,5,6,7])
arr7 = np.array([5,6,7,8,4,3])
np.setxor1d(arr6, arr7)
```

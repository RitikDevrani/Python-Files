# Data Type
- In Python, A data type is a classification that specifies the type of value a varible can hold. We can check a variable's data type using the type() function.

# Pytohon Data Types 
- A Data type defines what type of data a variable can hold and what operations can be performed on it.
   - Python commonly uses 7 main data types:
       - Numeric Data Types:
          - int - Integer numbers(whole numbers)
          - float - Decimal numbers
          - complex - Complex numbers(real+imaginary)
       - String(str)
          - Strores text, written inside quotes("" or '')
       - Boolean(bool)
          - Stores only True or False values.
       - List
          - Stores multiple items in an ordered collection, written inside square brackets [].
       - Tuple
          - Stores multiple items in an ordered but immutable collection (cannot be changed), written inside parentheses ().
       - Set
          - Stores unique items in an unordered collection, written inside curly braces {}.
       - Dictionary (dict)
          - Stores data in key - value pairs.

```python
x = 54  # This is a integer data type
print(x)
type(x)
```

```python
price = 52.99  # float data type
print(price)
type(price)
```

```python
com = 8j  #Complex data type
print(com)
type(com)
```

```python
z = 2 + 3j     # complex data type
print(z.real)  # to print real number
print(z.imag)  # to print complex number
```

```python
Num = "786"  # String Data type
print(Num)
type(Num)
```

```python
adult = True  # Boolean Data Type
print(adult)
type(adult)
```

```python
bool(1)
```

```python
bool(0)
```

```python
my_list = ["Ritik", 22, "Python", 99.9, 4j, "Ritik"]  # List Data type
print(my_list)
print(my_list[3])
my_list.append("Devrani")
print(my_list)
```

```python
my_tuple = (2,"Ritik", 5, 5.33, 3j, "Ritik")  # Tuple data type
print(my_tuple)
type(my_tuple)
```

```python
my_set = {1, 2, 2, "Ritik", "Ritik", "ritik", "ritik", 4j, 60.99, 59.44} # Set Data type
print(my_set)
type(my_set)
```

```python
my_dict = {"Name" : "Ritik", "Course" : "Data Analyst", "Age" : 22} # Dictionary Data type
print(my_dict)
print(my_dict["Name"])  # Ritik
print(my_dict.keys())
print(my_dict.values())
type(my_dict)
```

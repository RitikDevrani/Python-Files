# What is a Dictionary in Python?

 - A Dictionary Python ka ek collection data type hai jo key-value pairs me data store karta hai.
Iska matlab har value ek unique key ke through access hoti hai.

# Definition:

 - A dictionary is an unordered, changeable (mutable) collection of items stored in key: value pairs, enclosed in curly braces { }.

# Syntax:
 - dict_name = {
     key1: value1,
     key2: value2,
     key3: value3
   }

```python
student = {
    "name": "Ritik",
    "age": 23,
    "course": "BCA"
}
print(student)
```

# Important Features of Dictionary:
  - Feature	 = Description
    - Key-Value Pair	= Every value is connected to a key
    - Keys must be unique	= No duplicate keys allowed
    - Mutable	= You can add, remove, update values
    - Unordered	= Items do not follow fixed order
    - Keys must be immutable	= int, string, tuple allowed; list not allowed

# Dictionary Methods

| Method                | Description                            | Example                     | Output                          |
| --------------------- | -------------------------------------- | --------------------------- | ------------------------------- |
| **get(key, default)** | Returns value; no error if key missing | `d={"a":1}; d.get("a")`     | `1`                             |
|                       |                                        | `d.get("b", "Not Found")`   | `"Not Found"`                   |
| **keys()**            | Returns all keys                       | `d={"a":1,"b":2}; d.keys()` | `dict_keys(['a','b'])`          |
| **values()**          | Returns all values                     | `d.values()`                | `dict_values([1,2])`            |
| **items()**           | Returns list of (key,value) pairs      | `d.items()`                 | `dict_items([('a',1),('b',2)])` |
| **update()**     | Adds or updates key-value pairs           | `d={"a":1}; d.update({"b":2})`     | `{"a":1,"b":2}`                          |
| **setdefault()** | Returns value; inserts default if missing | `d={"a":1}; d.setdefault("b",100)` | `100` and dict becomes `{"a":1,"b":100}` |
| **fromkeys()**   | Creates dictionary from keys              | `dict.fromkeys(['x','y'],0)`       | `{"x":0,"y":0}`                          |
| **pop(key)**  | Removes key & returns value | `d={"a":1,"b":2}; d.pop("a")` | `1` and dict → `{"b":2}` |
| **popitem()** | Removes last key-value      | `{"a":1,"b":2}.popitem()`     | `('b',2)`                |
| **clear()**   | Removes all items           | `d={"a":1}; d.clear()`        | `{}`                     |
| **copy()** | Returns shallow copy | `d={"a":1}; d.copy()` | `{"a":1}` |



# Examples of Dictionary Operations

- Accessing Values (using keys)

```python
student = {"name": "Ritik", "age": 23}

print(student["name"])
print(student["age"])
```

- Using get() method

```python
print(student.get("name"))    # Ritik
print(student.get("city"))    # None (no error)
```

- Adding New Key-Value Pair

```python
student["city"] = "Kashipur"
print(student)
```

- Updating Value

```python
student["age"] = 22
print(student)
```

Removing Items

  
 - pop(key) → removes key and returns value

student.pop("age")

 - popitem() → removes last inserted item

student.popitem()

 - del → delete specific key

del student["name"]

 -  clear() → empties dictionary

student.clear()

```python
employee = {
    "id": 101,
    "name": "Aman",
    "salary": 45000,
    "department": "IT"
}

print("Name:", employee["name"])
print("Salary:", employee["salary"])

# update salary
employee["salary"] = 50000

# add new key
employee["city"] = "Delhi"

# loop through dictionary
for k, v in employee.items():
    print(k, ":", v)
```

# List vs Tuple vs Set vs Dictionary – Full Comparison Table

| Feature / Property          | **List**                         | **Tuple**                        | **Set**                              | **Dictionary**                                      |
| --------------------------- | -------------------------------- | -------------------------------- | ------------------------------------ | --------------------------------------------------- |
| **Definition**              | Ordered collection of items      | Ordered and immutable collection | Unordered collection of unique items | Collection of items in key–value pairs              |
| **Syntax**                  | `[ ]`                            | `( )`                            | `{ }`                                | `{key: value}`                                      |
| **Order**                   | ✔ Ordered                        | ✔ Ordered                        | ❌ Unordered                          | ❌ Unordered (Python 3.7+ insertion order preserved) |
| **Indexing**                | ✔ Supports indexing              | ✔ Supports indexing              | ❌ No indexing                        | Access through keys only                            |
| **Mutable (changeable)**    | ✔ Yes                            | ❌ No                             | ✔ Yes                                | ✔ Yes                                               |
| **Duplicates allowed**      | ✔ Yes                            | ✔ Yes                            | ❌ No                                 | ✔ Keys: No, ✔ Values: Yes                           |
| **Data types supported**    | Heterogeneous                    | Heterogeneous                    | Heterogeneous                        | Keys: immutable, Values: any type                   |
| **Speed**                   | Slower than tuple                | Fastest                          | Faster for membership tests          | Fast for key-based lookup                           |
| **Use Case**                | When items need to change        | When data should not change      | When unique items are needed         | When key–value mapping is needed                    |
| **Methods Available**       | Many (append, insert, pop, sort) | Very few (count, index)          | Many (add, remove, union)            | Many (keys, values, items, update)                  |
| **Allows Nesting**          | ✔ Yes                            | ✔ Yes                            | ✔ Yes (must be immutable items)      | ✔ Yes                                               |
| **Memory Usage**            | More than tuple                  | Least memory                     | Moderate                             | Highest                                             |
| **Duplicate Keys Allowed?** | —                                | —                                | —                                    | ❌ No                                                |
| **Best Feature**            | Flexible & widely used           | Fast & secure (immutable)        | Unique items + fast membership       | Fast lookup using keys                              |
| **Example**                 | `[10, 20, 30]`                   | `(10, 20, 30)`                   | `{10, 20, 30}`                       | `{"a": 10, "b": 20}`                                |

# Encapsulation in Python

- Encapsulation is an Object Oriented Programming concept that refers to binding data (variables) and methods (functions) into a single unit (class) and restricting direct access to some of the object's components.


- Encapsulation means:
   - Wrapping data and methods together
   - Protecting data from unauthorized access



- Encapsulation = Data Hiding + Data Binding

# Why Encapsulation is Needed?

- Without encapsulation:

  - Anyone can change data directly

  - Data becomes unsafe

  - Program becomes difficult to maintain


- Encapsulation provides:
  - Security
  - Controlled access
  - Data integrity
  - Maintainability

# Real-World Example of Encapsulation
 - Example: Bank ATM

 - You cannot access balance directly

 - You must use:

    - deposit()

    - withdraw()

    - check_balance()

- This is encapsulation.

# How Encapsulation is Implemented in Python?

- Python implements encapsulation using:

   - Classes

   - Access Modifiers

   - Getter and Setter Methods

# Access Modifiers in Python (Core Concept)

- Python has three types of access modifiers:

| Access Modifier | Syntax       | Meaning                                 |
| --------------- | ------------ | --------------------------------------- |
| Public          | `variable`   | Accessible everywhere                   |
| Protected       | `_variable`  | Accessible within class and child class |
| Private         | `__variable` | Accessible only within same class       |


- Public Members
  - Public members are accessible from anywhere.
  - No restriction
  - Least secure

```python
class Student:
    def __init__(self, name):
        self.name = name   # public variable

    def show(self):
        print(self.name)

s = Student("Ritik")
print(s.name)      # accessible
s.show()
```

- Protected Members
  - Protected members are intended to be accessed only inside class and its child class.
  - Note: Python does not strictly enforce protection. It is a convention.
  - Used in inheritance
  - Medium level security

```python
class Employee:
    def __init__(self, salary):
        self._salary = salary   # protected variable

class Manager(Employee):
    def show(self):
        print(self._salary)

m = Manager(50000)
m.show()
```

```python
m._salary
```

- Private Members
   - Private members are accessible only inside the same class.
   - Data hidden
   - High security

   - Syntax:
      - __variable


> Python performs Name Mangling:

     __balance → _ClassName__balance


```python
class Bank:
    def __init__(self):
        self.__balance = 5000   # private variable

    def show_balance(self):
        print(self.__balance)

b = Bank()
b.show_balance()

# print(b.__balance)  Error
```

- Name Mangling (Internal Working)

   - Python internally changes:
      - __balance → _Bank__balance


- Proof

```python
print(b._Bank__balance)   # possible but not recommended
```

```python
b._Bank__balance = 10000
```

```python
print(b._Bank__balance) 
```

- Encapsulation is logical, not strict like Java.

# Getter and Setter Methods
> Getter and Setter methods are special methods used to access (get) and modify (set) the value of private variables in a controlled and secure way.


- Getter → reads the value of a variable

- Setter → updates the value of a variable

- Why They Are Needed?
   - Private variables cannot be accessed directly outside the class.
   - Getter and setter provide controlled access.

- Relationship with Encapsulation
   - Encapsulation principle:

      - Data Hiding + Controlled Access = Encapsulation


      - Getter & Setter provide controlled access.

```python
class Student:
    def __init__(self):
        self.name = "ritik"
        self._course = "Data Analysis"
        self.__marks = 500   # private variable

    def get_marks(self):
        return self.__marks

    def set_marks(self, value):
        if value >= 0:
            self.__marks = value
        else:
            print("Invalid Marks")
s = Student()
s.set_marks(85)
print(s.get_marks())
```

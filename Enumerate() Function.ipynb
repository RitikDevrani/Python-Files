# Enumerate Function 
- enumerate() is a built-in function of python, it provide index number with every element of a iterable like list, tuple, string.


- When we apply loop on any list, tuple, string or any sequence then enumerate() automatically return both Index(by default = 0) and Value.


- Syntax

  - enumerate(iterable, start=0)
    - iterable = list, tuple, string.
    - start = Index starting number (default = 0). 

- Example - Enumerate on List

```python
fruits = ["apple", "mango", "banana"]
for index, item in enumerate(fruits):
    print(index, item)
```

- Example - Start index from custom number

```python
color = ["red","green","blue","black"]
for index, item in enumerate(color, start=101):
    print(index, item)
```

- Example - Using enumerate with string

```python
string = "Python"
for index, item in enumerate(string, start=1):
    print(index, item)
```

- Example - Enumerate inside list

```python
names = ["Ritik","Rahul","Ankit","Aman"]
print(list(enumerate(names)))
```

- Example - Enumerate with Tuple

```python
fruits = ("apple", "mango", "banana")
for index, item in enumerate(fruits):
    print(index, item)
```

- Example in a Practical situation.
   - when we need index of a special item.

```python
numbers = [10,20,30,40,50]
for i , num in enumerate(numbers):
    if num == 30:
        print("30 found at",i)
```

- Example - create Dictionary using enumerate.

```python
fruits_ = ["apple", "mango", "banana", "cherry"]
result = dict(enumerate(fruits_))
print(result)
```

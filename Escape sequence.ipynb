# Escape Sequence in Python
- Escape sequence are  special character combination that begin with a backslash (\) and are used to represent characters that are difficult or impossible to type directly (like newline, tab, quotes, etc). 

- \n - New line

```python
 print("Hello\nWorld")
```

- \t - Tab space

```python
print("Hello\tWorld")
```

- \\- Double backslash 

```python
print("c:\\newfolder\\documents")
```

-    \' - single quotes 

```python
print('Hello it\'s me ritik')
```

-   \" - Double quotes 

```python
print("Hello \"How are you?\"")
```

- \r - carriage return

```python
print("Ritik\rdev")
```

```python
print("Python\rjava")
```

```python
print("Hello\rHi")
```

```python
a = [3,5,7,3,7]
b = [4,6,7,2,7]
print(a,"\n",b)
```

```python
 print("Hello\aWorld")
```

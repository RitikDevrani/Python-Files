# filter() function in Python
- filter() ek built-in function hai jo given function ko har element par apply karke sirf woh elements return karta hai jinke liye function truthy value (True) return kare.
- Python 3 me filter() lazy iterator return karta hai (map ki tarah).


- 🔹 Syntax
    - filter(function, iterable)

       - function: ek callable jo ek element ko argument lega aur True/False (ya truthy/falsy) return karega.

       - Agar function None diya gaya hai → filter() iterable ke elements me se jo truthy hain unko return karega.

       - iterable: list, tuple, string, set, generator, etc.


- 🔸 Return type

   - Python 3: filter() filter object (an iterator) return karta hai — isse list() / tuple() / set() se convert kar sakte ho.

   - Python 2: filter() list return karta tha (old behavior).

- even numbers with filter function

```python
def is_even(n):
    return n % 2 == 0

nums = [1,2,3,4,5,6]
result = filter(is_even, nums)
print(result)           # <filter object ...>
print(list(result))     # [2, 4, 6]
```

Explanation: is_even True return karta hai jab number even ho; filter sirf unhi ko pass karta hai.

- Example using lambda with filter

```python
nums = [10, 15, 20, 25, 30]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)    # [10, 20, 30]
```

- filter() Example with strings — vowels 

```python
chars = "abcdefghijklmnopqrstuvwxyz"
vowels = list(filter(lambda ch: ch in "aeiou", chars))
print(vowels)   # ['a', 'e', 'i', 'o', 'u']
```

- filter(None, iterable) — remove falsy values

```python
data = ["", "Ritik", None, " ", 0, 25, [], [1]]
cleaned = list(filter(None, data))
print(cleaned)  # ['Ritik', ' ', 25, [1]]
```

Note: " " (space) is truthy, "" empty string is falsy

- Using on dict / keys / values

```python
d = {"a": 10, "b": 0, "c": 5}
# filter keys whose values > 0
keys = list(filter(lambda k: d[k] > 0, d))
print(keys)   # ['a', 'c']
```

- Using generators (shows laziness)

```python
def gen():
    for i in range(1, 6):
        print("producing", i)
        yield i

filtered = filter(lambda x: x % 2 == 1, gen())
print("filter created")
print(list(filtered))
```

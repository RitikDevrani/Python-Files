# 5. What is floor division operator? Give an example with output.

- Floor Division Operator in Python (//)

Floor Division operator // ka use division karne ke baad result ko lowest integer (floor value) me convert karne ke liye hota hai.

- Definition:

Result always becomes the largest integer that is LESS than or equal to actual division result.

- Example 
print(10 // 3)

Output:
3


Explanation:
10 ÷ 3 = 3.333…, floor value = 3

```python
print(10 // 3)
print(10 / 3)
```

# Indexing in Python
- Indexing means accessing element of a sequence (like string, list, tuple) using their position number(index)
   - Python uses zero-based indexing - first element = index 0.
   - Negetive indexing also work - last element = index -1.

```python
my_string = "Python is easy"
print(my_string[3]) 
print(my_string[-4])
```

```python
print(my_string[6])
```

```python
my_list = [3, "Ritik", 3.5, 5j]
print(my_list[3])
print(my_list[-3])
```

# Slicing in Python
- Slicing is used extract a part of a sequence (string, list, tuple) using:
- sequence[start: stop: step]
  - start - index where slice begins (default=0)
  - stop - index where slice ends (exclusive)
  - step (stride) - how much to move afters each element (default = 1).

```python
my_list1 = [3, "Ritik", 3.5, 5j, 5, "Dev"]
my_list1[:]
```

```python
my_list1[0:3]
```

```python
my_list1[2:6]
```

```python
my_list[6]
```

# Stride -- Long Steps
- String slicing can also accept a third parameter, the stird, which refers to many steps you want to take from the firs character of the string.
  - Sride = Step of the string.
  - It tells Python how many positions to skip when moving through the sequence.
- Syntex
  - sequence[start: stop: stride]

```python
string = "Hello how are you"
print(string[4])
print(string[0:])
print(string[0::3])
print(string[0::7])
```

```python
my_list_1 = ["how", 23, 3.5, 6j, True]
print(my_list_1[0::3])
```

```python
my_tuple_1 = ["how", 23, 3.5, 6j, True]
print(my_tuple_1[2::2])
```

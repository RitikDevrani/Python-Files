# Input Function

 - The input() function is used to take input from the user during program execution.
It always returns the input as a string (text), even if the user enters numbers.



 - Syntax:
   - variable = input(prompt)

   - prompt → (Optional) A message shown to the user before taking input.

   - variable → The value entered by the user is stored in this variable.

```python
Name = input("Enter your Name") # Basci example of input function
print("Your Name is : ",Name)
```

```python
age = int(input("Enter your age : "))  # Taking numeric input
print("Your are ",age,"year old")

price = float(input("Enter the price of a pen"))
print("Price of a pen is ",price,"rupees.")
```

```python
num1 = int(input("Enter first number: "))  # Performing Calculations with Input
num2 = int(input("Enter second number: "))
sum = num1 + num2
print("The sum is:", sum)
```

- If you don’t convert inputs to int, the result will concatenate strings instead of adding numbers:

```python
# Wrong example
a = input("Enter a: ")
b = input("Enter b: ")
print(a + b)  # If a=5, b=7 → Output: 57 (string concatenation)
```

```python
# Introduction Program

# Name
while True:
    Name = input("Enter your name: ")
    if Name.isalpha():  # Checks if all characters are alphabets
        break
    else:
        print("❌ Please enter a valid name (only letters).")

# Address
Address = input("Enter your address: ")

# Age
while True:
    try:
        Age = int(input("Enter your age: "))
        if Age >= 18:
            break
        else:
            print("❌ Age must be 18 or above.")
    except ValueError:
        print("❌ Please enter a valid number for age.")

# College Stream
while True:
    College_stream = input("Enter your stream: ")
    if College_stream.isalpha():
        break
    else:
        print("❌ Please enter a valid stream (only letters).")

# College Name
College_name = input("Enter your college name: ")

# Percentage
while True:
    try:
        percentage = int(input("Enter your percentage: "))
        if 0 <= percentage <= 100:
            break
        else:
            print("❌ Percentage must be between 0 and 100.")
    except ValueError:
        print("❌ Please enter percentage in digits only.")

# CGPA
while True:
    try:
        cgpa = float(input("Enter CGPA you got: "))
        if 0 <= cgpa <= 10:
            break
        else:
            print("❌ CGPA must be between 0 and 10.")
    except ValueError:
        print("❌ Please enter CGPA in decimal format (example: 8.9).")

# Profession & Hobbies
profession = input("Enter your profession: ")
hobbies = input("Enter your hobbies: ")

# Final Output
print("\n🎓 --- Personal Introduction --- 🎓")
print(f"Hello everyone! My name is {Name}, and I am from {Address}.")
print(f"I am {Age} years old. I have done my college in {College_stream} from {College_name}.")
print(f"I have scored {percentage}% with a CGPA of {cgpa}.")
print(f"My profession is {profession}. My hobbies are {hobbies}.")
```

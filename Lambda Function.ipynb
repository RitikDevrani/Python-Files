# Lambda Function
 - Lambda is an anonymous function of python means a Function without name.
 - It is used to create short, simple, one-line function.
     - Fast
     - short
     - Inline Defined
     - No def keyword needed

- Syntax of Lambda Function
   - lambda arguments: expression

- Explaination 
  - arguments - input parameters
  - expression - Single expression whose result is returned automatically.

  - Note - lambda function does not contain return keyword - return is implicity.

- Example : add two numbers

```python
add = lambda a,b : a+b
print(add(5,9))
```

```python
mul = lambda x,y,z : x*y*z
print(mul(5,6,3))
```

```python
square = lambda x : x*x  # square a number
print(square(5))
```

```python
check = lambda a : "Even"if a%2 == 0 else "Odd"
print(check(9))
```

```python
A = list(map(lambda x:x*5, [4, 8, 2, 6, 2]))
print(A)
```

```python
A = list(map(lambda x:x%2==0, [4, 8, 2, 6, 2,3,9,3,7,4,8,3]))
print(A)
```

```python
num = [4, 8, 2, 6, 2,3,9,3,7,4,8,3]
even = list(map(lambda x : x%2 == 0, num))
print(even)
```

```python
no = [2,3,4,5,6,74,76, 63, 73]
even = tuple(map(lambda j:j*3, no))
print(even)
```

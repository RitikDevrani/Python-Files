# List
- A list in Python is a verstile and widely-used type allows you to store multiple items in a single variable. It is one of Python's built-in data structure and offers flexiblity in handling collections of data.
- List is a collection of different data types like integer, Float, String as well as Objects.
- Each element or value of the list is called item, We can also say that list is a Sequence of items, it is a sequence data type.
- A list is created within square brackets [].

  My_list = [integer, float, string, boolean]

# Key Features of List
- Ordered: List maintain the order of element. The first element has an index of 0, the second 1, and so on.
- Changeable(Mutable): You can modify, add, or remove element after the list is created.
- Allows Duplicates: List can contain duplicate values.
- Hetrogeneous: A list can store element of different data types(eg., integer, strings, float, etc)

# Syntax:
- list_name = [element1, element2, element3, .....]

```python
My_list = [1, 44.77, "Ritik", True]
print(My_list)
type(My_list)
```

# Properties of a list
- A List allows duplicate items(Values) in it.

```python
list1 = [1, 1, 2, 2, 3.5, 3.5, "Ritik", "Ritik", 4+5j]
print(list1)
```

- A list is mutable. It means a list items can be changed after creating the list.

```python
# append() - To enter a new value at the end of a list.
list1.append("rahul")
print(list1)
```

```python
# insert() - To insert a new value at the given index of a list.
list1.insert(2,13)  # (index no, new value)
list1
```

```python
# index() - To show the index of a value.
list1.index(4+5j)
```

```python
# remove() - To remove a value from the list
list1.remove("rahul")
print(list1)
```

```python
# pop() - To delete a vlaue from a particular index (default last if index no not given)
list1.pop(8)
print(list1)
```

```python
nested_list = [3, 5, 4.3 , ["ritik", "humans"], 5, 9]
print(nested_list)
print(nested_list[3][1])
print(nested_list[3][0])
```

# List Methods

| Method    | Purpose                                |
| --------- | -------------------------------------- |
| append()  | Add element at end                     |
| insert()  | Add element at specific index          |
| extend()  | Add multiple elements                  |
| remove()  | Remove first occurrence of value       |
| pop()     | Remove by index (returns removed item) |
| clear()   | Remove all items                       |
| index()   | Return index of first occurrence       |
| count()   | Count occurrences                      |
| sort()    | Sort list ascending/descending         |
| reverse() | Reverse list                           |
| copy()    | Make a copy                            |


```python
str_list = ["apple", "mango", "banana"]
num_list = [4,76,2,7,9,8,3,9]
mixed_list = [4,76,2,7,9,"apple", "mango",8,3,9,"banana"]
print(str_list, num_list, mixed_list)
```

```python
#append()
str_list.append("orange")
num_list.append(18)
mixed_list.append("papaya")
print(str_list,"\n",num_list,"\n", mixed_list)
```

```python
#insert
str_list.insert(3,"cherry")
num_list.insert(4,18)
mixed_list.insert(2,"orange")
print(str_list,"\n",num_list,"\n", mixed_list)
```

```python
#extend
str_list.extend(["papaya","cherry"])
num_list.extend([45,90])
mixed_list.extend([2,"orange"])
print(str_list,"\n",num_list,"\n", mixed_list)
```

```python
str_list.extend("papaya")
print(str_list)
```

```python
#remove
str_list.remove("cherry")
num_list.remove(18)
mixed_list.remove("orange")
print(str_list,"\n",num_list,"\n", mixed_list)
```

```python
#pop
str_list.pop(4)
str_list.pop() #default last value remove
num_list.pop(3)
mixed_list.pop(6)
print(str_list,"\n",num_list,"\n", mixed_list)
```

```python
#sort
str_list.sort()
num_list.sort()
#mixed_list.sort() - wrong numbers and strings can not be sorted together
print(str_list,"\n",num_list,"\n", mixed_list)
```

```python
#revers
str_list.reverse()
num_list.reverse()
mixed_list.reverse() 
print(str_list,"\n",num_list,"\n", mixed_list)
```

```python
#count
print(str_list.count("p"))
print(num_list.count(9))
print(mixed_list.count("apple")) 
# print(str_list,"\n",num_list,"\n", mixed_list)
```

```python
#index
print(str_list.index("papaya"))
print(num_list.index(18))
print(mixed_list.index("apple")) 
# print(str_list,"\n",num_list,"\n", mixed_list)
```

```python
print(id(str_list))
print(id(num_list))
print(id(mixed_list))
```

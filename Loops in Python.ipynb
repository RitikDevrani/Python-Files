# Loops in Python : 
- Loops allow us to execute a block of code multiple times without writing it again and again.
- They are used to perform repetitive tasks efficiently until a specific condtion.


  - Types of Loops
     - While loop
     - For loop
     - Nested loop

```python
string = "Python is easy"
print(string[0])  #without loop
print(string[1])
print(string[2])
print(string[3])
print(string[4])
```

```python
for s in string: # with loop
    print(s)
```

# 1. While Loop :
- A while loop repeatedly executes a block of code as long as given condtion is True.
- it checks the condtion before each iteration.


   - Syntax
       - while conditon
       - code block to execute

```python
count = 1
while count < 5:
    print(count)
    count += 1
else:
    print("Loop Completed")
```

```python
while True:
    print("Hello Python!")  # infinite loop 
```

```python
num = 5
count_1 = 1
while count_1 <= 10:
    print(num*count_1)
    count_1 += 1
    
```

```python
num1 = int(input("Enter a number to get table"))
fixed_num = 1
while fixed_num <= 10:
    print(num1,"*",fixed_num,"=",fixed_num*num1)
    fixed_num += 1
```

# For Loop
- A for loop in Python in used to iterte over a sequence(like a list, tuple, dictionary, set or range) and execute a block of code for each element in that sequence.

    - Syntax
       - for variable in sequece
       - code block to execute.

```python
for i in range(1,6):
    print("*")
```

```python
python = "Python is easy"
for p in python:
    print(p)
```

```python
python = "Python is easy"
for p in python:
    print(p,end="")
```

# Nested Loop
- A nested loop means a loop inside another loop.
- The outer loop runs first.
- For every single iteration of the outer loop, the inner loop runs completely.
- Nested loops are used when we need repeated actions inside inside another repeated action - like printing patterns, table, matrices, etc.

```python
a = int(input("Enter a number :"))
for i in range(1, a+1):
    for j in range(1,11):
        print(i,"*",j,"=",i*j)
    
```

```python
for i in range(5,0,-1):
    for j in range(1,2):
        print("*"*i*j)

for i in range(1,6):
    for j in range(1,2):
        print("*"*i*j)


        
```

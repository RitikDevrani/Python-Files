# map() Function in Python

 - map() ek Python ka built-in function hai jo kisī function ko har element par apply karta hai aur result return karta hai.

- Syntax:

   - map(x,y)
   - map(function, iterable)
   - map(function, sequence\values)


     - function → jo operation karna hai

     - iterable → list, tuple, set, etc.

     - map() lazy object return karta hai jisko list/tuple me convert karna padta hai.

# Important Points

  - map() har element par function apply karta hai.

  - map() list return nahi karta, ek map object deta hai → jisko list() se convert karna padta hai.

  - map() multiple lists ke sath bhi use ho sakta hai jab lambdas me multiple arguments ho.

```python
list(map(lambda x:x+1, [1,3,45, 6, 4,6])) #adding 1 to each element and converting into list
```

```python
tuple(map(lambda x:x+5, [1,3,45, 6, 4,6])) #adding 5 to each element and converting into tuple
```

```python
set(map(lambda x:x+10, [1,3,45, 6, 4,6, 4, 6,3,67,45])) #adding 10 to each element and converting into set
```

```python
A = list(map(lambda x:x*5, [4, 8, 2, 6, 2]))
print(A)
```

- Convert all items to string

```python
nums = [10, 20, 30]

result = map(str, nums)

print(list(result))
```

- Converting all string to uppercase

```python
names = ["Ritik", "pankaj", "anuj", "sandeep", "pawan"]
NAMES = map(str.upper, names)
print(list(NAMES))
```

- Add two lists element-wise

```python
a = [1, 2, 3]
b = [4, 5, 6]

result = map(lambda x, y: x + y, a, b)

print(list(result))
```

- Square of each element number

```python
numbers = [3, 5, 7, 3, 8, 9, 3.4, 4.7, 9.3]
square_num = map(lambda x:x*x, numbers)
print(list(square_num))
```

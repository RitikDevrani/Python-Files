# Mutable and Immutable Data types in Python
 - Mutable - Can be changed or modified after creation
 - Immutable - Cannot be changed once created 

# Mutable Data Types(Changeable)
- Mutable objects can be changed in place, without creating a new object.
- Mutable data type Can be changed or modified after creation with the same memory location.
  - List
  - Dictionary(dict)
  - Set

```python
list_1 = ["apple", "mango", 13, 4.6, 5j, True, False]
print(id(list_1))
print(list_1)
list_1.append("banana")
print(id(list_1))
print(list_1)
```

```python
list_2 = ["apple", "mango", 13, 4.6, 5j, True, False]
print(id(list_2))
print(list_2)
list_2.insert(3,20)
print(id(list_2))
print(list_2)
```

```python
list_2 = ["apple", "mango", 13, 4.6, 5j, True, False]
list_2[6] = True
print(list_2)
```

```python
set_1 = {"apple", "mango", 13, 4.6, 5j, True, False}
print(id(set_1))
print(set_1)
set_1.remove(False)
print(id(set_1))
print(set_1)
```

```python
set_2 = {"apple", "mango", 13, 4.6, 5j, True, False}
print(id(set_2))
print(set_2)
set_2.add(75)
print(id(set_2))
print(set_2)
```

```python
print(list_1[3])
print(list_1.index(13))
```

```python
dict_1 = {"Name":"Ritik", "age":22, "city":"Rudrapur", "per":75.5}
print(id(dict_1))
print(dict_1)
dict_1["Name"] = "Aman" #changeable
print(id(dict_1))
print(dict_1)
```

```python
print(dict_1)
dict_1["class"] = 2  #modified
print(id(dict_1))
print(dict_1)
```

```python
dict_ = {"Name":"Ritik", "age":22, "city":"Rudrapur", "per":75.5}
dict_.pop("Name")
print(dict_)
```

# Immutable Data Types(Not Changeable)
 - Immutable objects cannot be modified after the are created.
 - If you try to change them, a new object is created instead and memory location is changed with new object.
   - String
   - Tuple
   - Integer(int)
   - Float
   - Boolean

```python
String = "Hello guys good morning"
String[2] = "r"
```

```python
print(String)
```

```python
s_no = "20"
s_no_1 = "20"
print(id(s_no))
add = s_no + s_no_1
print(add)
print(id(add))
```

```python
my_tuple = (2, "Ritik", 3.7, 3j, )
my_tuple[2] = "Dev"
```

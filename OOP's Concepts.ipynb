# Magic Method
- A magic method is a special method in Python that is called automatically to perform built-in operations on objects,
  like creation, printing, addition, comparison, etc.

```python
class Student:
    def __init__(self,name):
        self.name = name

    def __str__(self):
        return f"Student name is {self.name}"
```

```python
x = Student(10)
y = Student(20)
x+y
```

```python
class Add:
    def __init__(self,num):
        self.num = num

    def __add__(self,other):
        return self.num + other.num
```

```python
a = Add(10)
b = Add(20)
print(a+b)
```

```python
class Equal:
    def __init__(self,num):
        self.num = num

    def __eq__(self,use):
        return self.num == use.num
```

```python
a = Equal(15)
b = Equal(15)
a == b
```

# Instance Method
- Methods in which instance variable and class variable are used.
- Used to access or modify both object and class state.
- It must have a self parameters to refer to the current object

```python
class Student:
    def __init__(self, name):
        self.name = name

    def show(self):
        print("Name:", self.name)

s1 = Student("Ritik")
s1.show()
```

```python
class Person:
    def __init__(self):
        self.person = []
        self.city = []
        self.person_d()

    def person_d(self):
        person = input("enter the person")
        city = input("enter the city")
        self.person.append(person)
        self.city.append(city)
        print(self.person, self.city)

Person()
```

# What is a Class Method
- A class method is a method that works with the class itself. not with individual objects. it takes cls the first parameter, which refers to the class.
- Class methods work on class variables, not object variables.

- Decorator
  - @classmethod

- Parameter
  - cls   -- represents the class

```python
class Student:
    college = "IGNOU"

    @classmethod
    def change_college(cls):
        cls.college = "DU"

print(Student.college)
Student.change_college()
print(Student.college)
```

```python
Student.college
```

- Explanation:

  - cls refers to class Student

  - Changes class-level data

  - No need to create object

# Static Method
- Similar to a normal function.
- It does not take the cls and self parameters
- It can not modify the state of the object or class.
- They behave like normal functions but placed inside a class for logical grouping.
   - Advantages
      - Safer Code - as they cannot change the class or instance variables.m

- Decorator
  - @staticmethod

```python
class Math:
    @staticmethod
    def add(a, b):
        return a + b

print(Math.add(10, 20))
```

# Comparison Table

| Feature              | Instance Method | Class Method   | Static Method   |
| -------------------- | --------------- | -------------- | --------------- |
| Decorator            | ❌               | `@classmethod` | `@staticmethod` |
| First Parameter      | `self`          | `cls`          | ❌               |
| Access Instance Data | ✔               | ❌              | ❌               |
| Access Class Data    | ✔               | ✔              | ❌               |
| Called By            | Object          | Class/Object   | Class/Object    |


```python
class Classmate:
    clg_name = "college"
    student1 = "Rahul"
    student2 = "Amit"
    student3 = "Pankaj"

    def __init__(self, subject, tname):
        self.subject = subject
        self.tname = tname

    @classmethod
    def Student_list(cls):
        print("Student first",cls.student1)
        print("Student second",cls.student2)
        print("Student third",cls.student3)
        

    def Student1(self, marks1):
        self.marks1 = marks1
        print("Student name :",Classmate.student1)
        print("College name :",Classmate.clg_name)
        print("Subject name :",self.subject)
        print("Teacher name :",self.tname)
        print("Student marks :",self.marks1)

    def Student2(self, marks2):
        self.marks2 = marks2
        print("Student name :",Classmate.student2)
        print("College name :",Classmate.clg_name)
        print("Subject name :",self.subject)
        print("Teacher name :",self.tname)
        print("Student marks :",self.marks2)

    def Student3(self, marks3):
        self.marks3 = marks3
        print("Student name :",Classmate.student3)
        print("College name :",Classmate.clg_name)
        print("Subject name :",self.subject)
        print("Teacher name :",self.tname)
        print("Student marks :",self.marks3)
         

    @staticmethod
    def grade(performance):
        print("Over all performance of Students :",performance)
    
```

```python
obj = Classmate("Python", "Mr. Sharma")

Classmate.Student_list()

obj.Student1(85)
obj.Student2(78)
obj.Student3(90)

Classmate.grade("Excellent")
```

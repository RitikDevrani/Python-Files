# Python Operators 
- In Python, operators are special symbols or keywords perform operations on values(operands). For example:
   - Arithmetic - Math
   - Comparison - Compare values
   - Logical - combine conditions
   - Assigment - Update variables
   - Bitwise - Binary operations
   - Membership - Check sequece membership
   - Identity - Check object memory identity

# Types of Python Operators
- 1. Arithmetic Operators
     - Used for mathematical calculations
       
       

```python
num_1 = 12
num_2 = 10
print(num_1 + num_2)
print(num_1 - num_2)
print(num_1 * num_2)
print(num_1 % num_2)
print(num_1 / num_2)
print(num_1 // num_2)
print(num_1 ** num_2)
```

- 3. Logical Operators
     - Used to combine conditional statement.
       
       

- 1. Logical AND (and)
      - Rule: Both conditons must be True - then result in True
      - If even one condition is False - result is False

```python
5<2 and 6<8
```

```python
no1 = 15; no2 = 20
no1<=no2 and no1<no2
```

```python
no1>=no2 and no1>no2
```

```python
no1<no2 and no1>no2
```

- 2. Logical OR(or)
     - Rule: If at least one condition is True - Rusult is True
     - Only when both are false - result is False

```python
a = 10; b = 12
a<b or b>a
```

```python
a<b or b==a
```

```python
a>b or b<a
```

# What is Polymorphism?

- The word Polymorphism comes from:

   - Poly → many

   - Morph → forms


- Polymorphism means one name, many forms.

- In Python, polymorphism allows the same method, function, or operator to behave differently in different situations.

# Why Polymorphism is Needed?

- Polymorphism helps to:
   - Write flexible code
   - Reduce complexity
   - Improve code reusability
   - Follow real-world behavior

# Real-World Example
| Action | Different Forms  |
| ------ | ---------------- |
| Speak  | Human, Dog, Baby |
| Draw   | Circle, Square   |
| Start  | Car, Bike        |

- Same action name, different behavior.

# Types of Polymorphism in Python

- Python supports polymorphism in four main ways:

  - Function Polymorphism
  - Method Overriding
  - Operator Overloading

# 1. Function Polymorphism

- Same function name works with different data types.

```python
print(len("Python"))      # string
print(len([1, 2, 3]))     # list
print(len((4, 5, 6, 7)))  # tuple 
# Same function, different objects
```

# 2. Method Polymorphism (Method Overriding)

- When child class provides its own implementation of a parent class method.

```python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

a = Animal()
d = Dog()
c = Cat()

a.sound()
d.sound()
c.sound()
                               # Same method name, different outputs
```

# 3.Operator Overloading
- Python allows operators to have different meanings.

```python
# Example: +
print(10 + 20)          # addition
print("Hello" + "Hi")  # string concatenation
print([1, 2] + [3, 4]) # list merging


# Same operator, multiple behaviors
```

```python
class Payment:
    def pay(self):
        pass

class UPI(Payment):
    def pay(self):
        print("Paid using UPI")

class Card(Payment):
    def pay(self):
        print("Paid using Card")

payments = [UPI(), Card()]

for p in payments:
    p.pay()
```

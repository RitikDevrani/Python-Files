# What is Python Module?
- A Module is a simple file that contains collections of functions, classes and variable and has a.py extension file. Eg - math, os, datetime.
   - We use module in our program by importing them.

- A module in Python is a file that contains Python code such as:

   - variables

   - functions

   - classes

   - executable statements

- A module is a Python file used to organize and reuse code.

# Why do we need Modules?

  - Break large programs into small manageable files

  - Reuse code in multiple programs

  - Improve readability

  - Improve maintenance

  - Avoid code duplication

# Types of Modules in Python

- Python has three main types of modules:

   - 1️. Built-in Modules
   - 2️. User-defined Modules
   - 3️. Third-party Modules

- 1️ Built-in Modules

    - These modules come pre-installed with Python.

    - Examples:

        - math

        - random

        - datetime

        - os

        - sys

```python
#Example: math module
import math

print(math.sqrt(25))
print(math.factorial(5))
```

- ️2. User-Defined Modules

   - Modules created by the programmer.

   - Step 1: Create a module file

   - name: mymodule.py

- Step 2: Use the module in another file
- import mymodule

- 3️ Third-Party Modules

- Modules created by other developers and installed using pip.

- Examples:

   - numpy

   - pandas

   - matplotlib

   - requests

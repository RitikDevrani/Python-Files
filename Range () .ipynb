# range() Function
- The range function creates a sequential list of numbers
- by defaul - this list starts from 0 ends at one value less than the specified number, and the list is incremented by 1
- The call to `list` is necessary because`range` by itself create a `range object` , and this must be converted to a list.

```python
range(5)
```

```python
a = range(5)
print(a)
```

```python
b = list(range(5))
print(b)
```

- if the `range` is called with one argument it produces values starting from 0

```python
c = list(range(51))
print(c)
```

- if the `range` is called with two arguments then it provides values from the first to the second argument

```python
x = list(range(1,16))
print(x)
```

```python
y = list(range(15,46))
print(y)
```

```python
range(20) == range(0,20)
```

```python
range(0,11) == range(10)
```

- range withe three agruments
- range(starting,endding number,steps/stride)


```python
p = list(range(5,85,5))
print(p)
```

```python
p = list(range(1,33,3))
print(p)
```

- example reverse a range

```python
num = list(range(10,0,-1))
print(num)
```

```python
num = list(range(38,0,-4))
print(num)
```

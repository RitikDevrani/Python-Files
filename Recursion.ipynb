# Recursion:
> Recursion is a programming technique where a function solves a problem by calling itself with a smaller input until if reaches a base condition.

```python
def demo():

    demo()
demo()
```

```python
i = 0
def demo():
    global i
    i += 1
    print("Hello World!",i)
    demo()
    print(i)
demo()
```

```python
n = int(input("Enter the num: "))
def natural(n):
    if n == 0:
        return
    print(n)
    return natural(n-1)
natural(n)
```

```python
def fact(n):
    if n == 0:
        return 1
    return n*fact(n-1)
fact(5)
```

```python
# Factorial Number:
def fact(n):
    if n >= 1:
        return n*fact(n-1)
    else:
        return 1
num = int(input("Enter the number: "))
if num <= 0:
    print("Please enter the positive number \"Try again\"")
else:
    value = fact(num)
    print(value)
    
    
```

# Advantages
- Makes code simple and clean.
- Easily breaks big problems int samller ones.
- Some problems are naturally recursive.
- Reduces repetitive code.

# Disadvantages
- Uses more memory.
- Slower than loops.
- Risk of stack overflow.
- Hard to debug.
- Loops are often more efficient.

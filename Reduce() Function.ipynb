# reduce() Function
- it is pythonp's functional programming function which reduce the  element values of a sequence into one vlaue.
- this function is in functool module.

- Mathematical interpretation
  - apply one function on pairs of value.
  - then apply the next value on result until list ends.
  - Final Output: single accumuted value.

- Syntax
   - from functool import reduce
   - ruduce(function, iterable, initializer)
   - `function` - operation/logic which will be applied on every pair of vlaue.
   - `iterable` - List, Tuple, etc.
   - `initializer`(Opetional) - Starting value(default=first item) 

- How reduce() works Internally
   - reduce(f[a,b,c,d])
     - Internally it does
     - Step1 : f(a,b)
     - Step2 : f(result, c)
     - Step3 : f(result, d)
     - Final result - returned 

- Example - sum of list elements

```python
from functools import reduce
nums = [3,4,5,8,4,9,6,4]
result = reduce(lambda a,b: a+b, nums)
print(result)
```

- Example multiply all numbers

```python
nums_1 = [2,3,4]
result_1 = reduce(lambda a,b: a*b, nums_1)
print(result_1)
```

- Find Maximum using reduce

```python
result_2 = reduce(lambda a,b: a if a>b else b, nums)
print(result_2)
```

- Find minimum using reduce

```python
result_3 = reduce(lambda a,b: a if a<b else b, nums)
print(result_3) 
```

- Reduce with initializer

```python
num_s = [1,2,5]
result_4 = reduce(lambda a,b: a+b, num_s, 10)
print(result_4)
```

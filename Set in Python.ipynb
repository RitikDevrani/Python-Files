# What is a Set in Python?

 - A set in Python is a built-in data type used to store unordered, unique, and mutable collections of elements. it is particularly useful when you need to eliminate duplicates or perform mathematical set operations like union, intersection, and difference.
 - A Set is a collection data type in Python which is used to store multiple unique items in a single      variable.
 - It is unordered, unindexed, and does not allow duplicate values.
 - items enclosed within curly braces { }.

# Syntax:
 - set_name = {item1, item2, item3, ...}

```python
my_set = {10, 20, 30, 40}
print(my_set) #The order may change because sets are unordered.
```

# Key Features of Set:
- Feature	         Description
  - Unordered	-    Elements don’t have a fixed position or index
  - Unindexed   -    You can’t access items using index numbers
  - No Duplicates -  Each element appears only once
  - Mutable     -	You can add or remove elements after creation
  - Heterogeneous - Can store different data types (int, str, float, etc.)

# Set Common methods

| Method        | Description                           | Example Code             | Output             |
| ------------- | ------------------------------------- | ------------------------ | ------------------ |
| **add()**     | Adds a single element                 | `s={1,2}; s.add(3)`      | `{1,2,3}`          |
| **update()**  | Adds multiple elements                | `s={1}; s.update([2,3])` | `{1,2,3}`          |
| **remove()**  | Removes element (error if missing)    | `s={1,2}; s.remove(2)`   | `{1}`              |
| **discard()** | Removes element (no error if missing) | `s={1,2}; s.discard(5)`  | `{1,2}`            |
| **pop()**     | Removes random element                | `s={10,20}; s.pop()`     | Random e.g. `{20}` |
| **clear()**   | Removes all elements                  | `s={1,2}; s.clear()`     | `set()`            |
| -------------------------- | ------------------- | ------------------------------- | --------- |
| **union()**                | Combine sets        | `{1,2}.union({2,3})`            | `{1,2,3}` |
| **intersection()**         | Common elements     | `{1,2,3}.intersection({2,3,4})` | `{2,3}`   |
| **difference()**           | A minus B           | `{1,2,3}.difference({2,3})`     | `{1}`     |
| **symmetric_difference()** | Elements not common | `{1,2,3}.sym_diff({3,4})`       | `{1,2,4}` |


```python
# Set with Duplicates
nums = {1, 2, 2, 3, 4, 4, 5}
print(nums) #will only print unique values
```

```python
fruits = {"apple", "banana"}
fruits.add("cherry") # Adding element
print(fruits)
```

```python
# Adding Multiple Elements (update)
fruits.update(["mango", "grapes"])
print(fruits)
```

```python
Set_s = {3,6,3,6,7,7,0,4,9}
Set_s.update("data","Analyst")  #withont [] brackets  
print(Set_s)
```

```python
# Removing Elements
fruits.remove("banana")    # Gives error if item not found
fruits.discard("apple")    # No error if item not found
print(fruits)
```

```python
print(fruits)
```

```python
fruits.pop()  # removes random item
print(fruits)
```

```python
fruits.clear() #removes all the items
print(fruits)
```

```python
set_a = {1,2,3,7,8,9}
print(set_a)
set_b = {4,5,6,1,2,3,7,8,9}
print(set_b)
print(set_a.union(set_b)) # combines sets and gives unique values
print(set_a.intersection(set_b)) # give common values
print(set_a.difference(set_b)) # seta - setb give remaining elements of a
print(set_a.symmetric_difference(set_b)) # give elements are not common
```

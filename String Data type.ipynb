# Python string data type

- A string in python a sequence of characters inclosed in single(''), double("") and triple quotes("""/""").


- Key features of string.
  - Immutable = Once a string is created, It can not be modified. Any operation that appears that actually creates a new string.
  - Unicode Support = Python string support unicode, allowing representation of characters from varios language and symbols.

```python
print("गुणवत्ता और संतुलन")
```

```python
print("😂")
print("👌")
```

```python
name = 'ritik'
print(name)
```

# Characteristics of String
- Python string ek sequence data type hai jisme characters ka collection hota hai.
Neeche important characteristics table ke form me diye gaye hain.

| **Characteristic**              | **Description (Hinglish)**                                         | **Example**                 | **Output**     |
| ------------------------------- | ------------------------------------------------------------------ | --------------------------- | -------------- |
| **1. Immutable (अपरिवर्तनीय)**  | String ko change nahi kar sakte; new string banani padti hai       | `s = "hello"; s[0] = 'H'`   | Error          |
| **2. Ordered / Indexed**        | Characters fixed order me rehte hain; index se access kar sakte ho | `"Python"[0]`               | `P`            |
| **3. Iterable**                 | For loop se har character ko iterate kar sakte ho                  | `for ch in "Hi": print(ch)` | H<br>I         |
| **4. Supports Slicing**         | Substring extract kar sakte hain using `[start:end]`               | `"Python"[0:3]`             | `Pyt`          |
| **5. Allows Duplicates**        | Same character multiple times aa sakta hai                         | `"aabbcc"`                  | `aabbcc`       |
| **6. Heterogeneous Characters** | Numbers, letters, symbols sab ho sakte hain                        | `"A1@#"`                    | `A1@#`         |
| **7. Dynamic Length**           | String ki length kitni bhi ho sakti hai                            | `len("Hello")`              | `5`            |
| **8. Unicode Support**          | Emojis, Hindi, any language support                                | `"नमस्ते😊"`                | नमस्ते😊       |
| **9. Concatenation Supported**  | `+` se strings join hoti hain                                      | `"Hello " + "Python"`       | `Hello Python` |
| **10. Repetition Supported**    | `*` se string repeat hoti hai                                      | `"Hi"*3`                    | `HiHiHi`       |


```python
s = "hello"; s[0] = 'H'  #Immutable = error dega
```

```python
"Python"[0]  #Ordered / Indexed = p
```

```python
"Python"[0:3]  # Supports Slicing
```

```python
"aabbcc"  #Allows Duplicates
```

```python
"A1@#"   #Heterogeneous 
```

```python
len("Hello")   #Dynamic Length
```

```python
"नमस्ते😊"  #Unicode Support
```

```python
"Hello " + "Python"    # Concatenation Supported
```

```python
"Ritik"*5    #Repetition Supported
```

```python
Name = "Ritik"      #Membership operator
"t" in Name
```

# Common String Methods in Python

#   1. upper()

String ko uppercase me convert karta hai.

```python
text = "hello"
print(text.upper())
```

```python
name = "ritik"
print(name.upper())
```

#  2. lower()

String ko lowercase me convert karta hai.

```python
print("PYTHON".lower())
```

```python
lang = "PYthOn"
print(lang.lower())
```

#  3. capitalize()

First letter ko capital banata hai.

```python
print("hello python".capitalize())

```

```python
cap = "hello python"
print(cap.capitalize())
```

#  4. title()

Har word ka first letter capital.

```python
print("python programming language".title())
```

#  5. strip()

Starting aur ending spaces remove karta hai.

```python
print("  hello  ".strip())
```

#  6. lstrip() & rstrip()

lstrip() → left side spaces remove

rstrip() → right side spaces remove

```python
print("  hello".lstrip())
print("hello  ".rstrip())
```

#  7. replace(old, new)

String ke part ko replace karta hai.

```python
print("Hello World".replace("World", "Python"))
```

#  8. split()

String ko list me tod deta hai (space ke basis par).

```python
print("Ritik Devrani".split())
```

```python
"apple,banana,mango".split(",")
```

```python
name = "Ritik devrani"
print(name.split())
```

# 9. join()

List ke elements ko join karta hai string ke sath.

```python
print("or".join(["a", "b", "c"]))
```

```python
my_name = ["Ritik", "Devrani"]
dash = "-"
dash.join(my_name)
```

```python
my_name1 = ["Ritik", "Devrani"]
"-".join(my_name1)
```

# 10. find()

Substring ki starting index return karta hai (not found → -1).

```python
print("Python".find("th"))
```

# 11. startswith()

Check karta hai string kis se start ho raha hai.

```python
print("Python".startswith("Py"))
```

# 12. endswith()

Check karta hai string kis se end ho raha hai.

```python
print("programming".endswith("ing"))
```

# 13. count()

Substring kitni baar aaya hai.

```python
print("banana".count("a"))
```

# 14. isdigit()
Check karta hai string me sirf digits hain ya nahi

```python
print("1234".isdigit())
```

```python
print("rahul".isalpha())   
print("rahul123".isalpha())
```

```python
print("rahul".isalnum())   
print("rahul123".isalnum())
print("346457".isalnum())   
print("rahul123@#$".isalnum())
```

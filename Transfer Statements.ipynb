# Python Transfer Statements:
- Transfer statement in Python are used to change the normal flow of a program.
- They help to skip, stop, or jump out of loops or blocks of code.

# 1. Pass Statement:
  - The pass statement is used as a placeholder in Python.
  - It does nothing, but it helps to avoid syntax errors when you want to keep the block empty temporarily. 

```python
for i in range(5): # It will give error
```

```python
for i in range(5): # it won't give error and pass the code to the next cell
    pass    # means it will avoid syntax error.
```

```python
for i in range(5):
    if i == 3:
        pass
    print(i)
```

```python
i = 1
while i <= 5:
    if i == 3:
        pass
    print(i)
    i += 1
```

# 2. Break Statement:
- The break stetement is used to exit a loop immediately, even if the loop condition is still true.


- Syntax.
  - for variable in sequence:
     - if condition:
       - break
     

```python
for i in range(10):
    if i == 4:
        print("Hello World")
        break
    print(i)
```

```python
for i in range(5):
    if i == 3:
        print(i)
        for j in range(5):
            print("Python is easy")
        break
    print("Hello World")
```

```python
i = 1
while i <= 10:
    if i == 6:
        break
    print(i)
    i += 1
```

```python
for i in range(8):
    if i == 5:
        print(i)
        break
```

# Continue Statement:
- The continue statement is used to skip the current iteration of the loop and move to the next iteration.


  - Syntax:
     - for i in sequence:
        - if condition:
          - continue

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
i = 1
while i <= 5:
    if i == 3:
        i += 1
        continue
    print(i)
    i += 1
```

```python
i = 5
while i >= 1:
    if i == 3:
        i -= 1
        continue
    print(i)
    i -= 1
```

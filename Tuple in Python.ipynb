# Tuple in Python 
- A Tuple is a collection data type in Python that is used to store multiple items in a single variable.
It is similar to a list, but the main difference is that tuples are immutable (i.e., unchangeable).
- A tuple is an ordered and immutable collection of items enclosed within parentheses ( ).

- Syntax:
  - tuple_name = (item1, item2, item3, ...)

```python
my_tuple = (10, 20, 30, 40)
print(my_tuple)
```

# Tuple Characteristics:
- Property & Description
  - Ordered	= Items have a defined order and can be accessed using an index
  - Immutable = You cannot change (add, update, or remove) items after creation
  - Allows Duplicates = Same value can appear more than once
  - Can store different data types = e.g., int, string, float, list, etc.

```python
tuple1 = (1, 2, 3, 4)
tuple2 = ("apple", "banana", "cherry")
tuple3 = (1, "hello", 3.5, True)
print(tuple1, tuple2, tuple3)
```

# Indexing/Accessing Tuple Elements

```python
fruits = ("apple", "banana", "cherry")
print(fruits[0])   # Output: apple
print(fruits[2])   # Output: cherry
```

```python
# Negative Indexing
numbers = (10, 20, 30, 40)
print(numbers[-1])   # Output: 40 (last element)
print(numbers[-2])
```

```python
data = (1, 2, 3, 4, 5)  # Printing tuple in range / Slicing Tuples
print(data[1:4])   # Output: (2, 3, 4)
```

# Tuple with One element
- You must add a comma after the element, otherwise Python treats it as a normal variable.

```python
single_tuple = (5,)     # Correct
not_tuple = (5)         # Wrong – treated as integer
```

- Tuple Operations

```python
tuple1 = (1, 2, 3)
tuple2 = (4, 5)

# Concatenation
result = tuple1 + tuple2
print(result)   # Output: (1, 2, 3, 4, 5)

# Repetition
repeat = tuple1 * 2
print(repeat)   # Output: (1, 2, 3, 1, 2, 3)
```

# Nested Tuple
- Tuples can also contain other tuples (nested tuples):

```python
nested = ((1, 2), (3, 4), (5, 6))
print(nested[1])      # Output: (3, 4)
print(nested[1][0])   # Output: 3
```

```python
# Tuple Example Program
student = ("Ritik", 22, "BCA", "Kashipur")

print("Student Name:", student[0])
print("Age:", student[1])
print("Course:", student[2])
print("City:", student[3])
```

# Type Casting 
- Type casting in Python to the process of converting a value from one data type to another.
- It is useful when performing operations between different data types or when formatting data in specific way.
- Python provides serral built-in functions for type casting.

- Common Functions
  - int() - Converts a value to an integer
  - float() - Converts a value to a floating-point number
  - str() - Converts a value to a string
  - list(), tuple(), set(), dic(), bool() - Convert values into respective data types

```python
name = "45" #String to integer
int(name)
print(name)
type(name)
```

```python
price = 50.99 #Float to integer
new_price = int(price)
print(new_price)
type(new_price)

```

```python
a = "56"
b = int(a)+45
print(type(a))
print(b)
type(b)
```

```python
x = "34d" # can not change string alphabets into other numeric data type
int(x)
```

# Arguments in a Function
- An argument is a value you pass into a function so the function can use it.

```python
def info(x,y):  # x , y are parameters
    print(x,y)
```

```python
info(5, "6") # 5, "6" are arguments
```

- Types of Arguments
  -  1. Positinal Arguments
     - Order is important

```python
def cal(a,b,c,d):
    add = a+b
    mul = a*b
    diff = c/d
    ex = c**d
    return (add, mul, diff, ex)
```

```python
total = cal(3,5,8,6)
print(total)
```

- 2. Keyword Argument
     - Order does not important

```python
def info(name, age, gender):
    print("Name :", name)
    print("Age :", age)
    print("Gender", gender)
```

```python
info(gender="Female", name="Riya", age=23)
```

- 3. Default Arguments
     - Function used a default value is provided.

```python
def car_details(Name="Nano", average=3):
    print(Name, average)
car_details()
```

```python
car_details()
```

```python
car_details("BMW",5)
```

# Variable-Length Arguments
 - *args is used when you want to pass positional values to a function in a (tuple).
 - (**)kwargs is used when you want to pass many named arguments(key-value pairs). 

```python
def multi(*args):
    print("Sum of values",sum(args))
    
```

```python
multi(4,6,3,7,3,8)
```

```python
def add(*number):
    print(number)
    sum = 10
    for i in number:
        sum += i
    print(sum)
add(12,13,14,11,2,3,4)
```

```python
def student_record(**data):
    print("Student Record")
    for key , value in data.items():
        print(key,":",value)
```

```python
student_record(Name = "Riya", Roll_no = 12, Class = "10th", Subject = "Math")
```

```python
def my_task(**values):
    print("This is my ridding details:",values)
    for key in values:
        print("My Task Name :",key)
        val_num = values[key]
        print("My task values_number :",val_num)
        print(f"{key}:{val_num}")
```

```python
my_task(Riding_no = 1, Speed = 112, Cutting = 75)
```

# What is nonlocal?
- The nonlocal keyword is used inside nested function(a function inside another function).
- It refers to a variable in the nearest outer(enclosing) function scope, not globla.
- You can modify the variable of the outer function from the inner function.

```python
def outer():
    x = 95
    print(x)
    def inner():
        x = 85
        print(x)
    inner()
    print(x)
outer()
```

```python
def outer():
    x = 95
    print(x)
    def inner():
        nonlocal x
        x = 85
        print(x)
    inner()
    print(x)
outer()
```

# advantages of Local Variable
> Safe - Cannot be changed accidentally by other functiona


> Ease Debugging - Errors are limited to the function


> Memory Efficient - Deleted after function execution


> Better readabilility - Code is clean and well-structured

# Disadvantages of Loval Varibles
> Limited Access - Cannot be used outside the function.

> Data sharing is difficult - Need return value of parameters

# Advantages of Global Variable
> Easy Access - Can be used by all functions


> Useful for shard data - Same value used in multiple functions

> Less parameter passing - No need to pass value again and again.

# Disadvantages of Gloabla Variables
> Risky - Any function can change the value.


> Hard Debugging - Difficult to track where value changed.

> Poor Readability - Makes code confusing in large programs.

> Not Secure - Data can be modified accidentally.

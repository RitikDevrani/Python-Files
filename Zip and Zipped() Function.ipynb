# Zip Function 
- zip() is a built-in function which combines multiple iterables(list, tuple, string) into element-wise in pair/tuple.


  - Syntax
     - zip(iterables)
     - iterables - means we can pass multiple iterables.

- Default Behaviour(length mismatch)- a length of a zip is always equal to shortest iterable's length.
- means extra element of longer iterables are ignored.

- Example two lists

```python
names = ["Ritik","Aman","Sandeep","Pankaj"]
Age = [22, 33, 19, 25]
Z = zip(names, Age)
print(list(Z))
```

- Example three iterables

```python
a = [1,2,3]
b = ['x', 'y', 'z']
c = [True, False, True]

a_b_c = zip(a,b,c)
print(list(a_b_c))
```

- Lengths different (Shortest wins)

```python
a = [1,2,3]
b = ['x', 'y', 'z']
c = [True, False]
d = [20, 22, 25, 25]

combine = zip(a,b,c,d)
print(list(combine))
```

# zip(*zipped) = Unzipping in python
- zip(*zipped) is used to unzip combined pair/tuple.
- means it reverse what zip() does

```python
# example
pair = [(1,"x", True),(2, "y", False),(3, "z", False),(4, "w", True)]
num, xx, boolean = zip(*pair)
print(num)
print(xx)
print(boolean)
```

- Example: Practical pipeline
   - suppose CSV rows read as list of tuple and need column

```python
rows = [
    ("id1", "Ritik", 22),
    ("id2", "Pawan", 24),
    ("id3", "Mohit", 26),
    ("id4", "Ankit", 25),
    ("id5", "Karan", 28),
]
ids, name, age = zip(*rows)
print(ids)
print(name)
print(age)
    
    
```

- Python has a simple syntax similar to english language, designed to be easy as well as fun.
- Python is easy to understan language that feels like reading simple english. This persue code nature to python makes it easy to learn and understandable for beginners.
- Python is an HIGH LEVEL general purpose PROGRAMMING LANGUAGE and is DYNAMICALLY TYPED, which works with OR in IDE(Integrated Development Environment).
- WRITE LESS, DO MORE : Python has syntax that allows developers to write programs with fewer lines than some other programming language. That means prototyping can be very quick.
- Python provides larger developer community, or open-source community for strong, supportand helpful documentation.
- It offers a wide range of libraries that can be used across various applications. For example:Numpy, SciPY, Keras and Seaborn.
